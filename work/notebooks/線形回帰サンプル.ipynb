```julia
import Pkg
using Flux
using Plots
```

```julia
x = hcat(collect(Float32, -3:0.1:3)...)
```

```julia
f(x) = @. 3x + 2
```

```julia
y = f(x)
```

```julia
x = x .* reshape(rand(Float32, 61), (1, 61))
```

```julia
plot(vec(x), vec(y), lw = 3, seriestype = :scatter, label = "", title = "Generated data", xlabel = "x", ylabel= "y")
```

```julia
flux_model = Dense(1 => 1)
```

```julia
function flux_loss(flux_model, features, labels)
    y_hat = flux_model(features)
    Flux.mse(y_hat, labels)
end
```

```julia
flux_loss(flux_model, x, y)
```

```julia
flux_model.weight, flux_model.bias
```

```julia
function train_model!(f_loss, model, features, labels)
    dLdm, _, _ = gradient(f_loss, model, features, labels)
    @. model.weight = model.weight - 0.0001 * dLdm.weight
    @. model.bias = model.bias - 0.0001 * dLdm.bias
end
```

```julia
for i = 1:1000000
    train_model!(flux_loss, flux_model, x, y)
end
```

```julia
flux_model.weight, flux_model.bias
```

```julia
flux_model.weight[1], flux_model.bias[1]
```

```julia
plot(reshape(x, (61, 1)), reshape(y, (61, 1)), lw = 3, seriestype = :scatter, label = "", title = "Simple Linear Regression", xlabel = "x", ylabel= "y")
plot!((x) -> flux_model.bias[1] + flux_model.weight[1] * x, -3, 3, label="linear-regression", lw=2)
```
